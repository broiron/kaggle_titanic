### Data Processing

```python
# Import Preproceesing Libs
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Drop columns that are not needed for modeling
train_data = train_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
test_data = test_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
# Handle missing values
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())
test_data['Fare'] = test_data['Fare'].fillna(test_data['Fare'].median())
train_data['Embarked'] = train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])
```

```python
# Scale the data
# train과 test 모두 동일한 전처리 -> 정규 분포의 모양을 띄도록 함 
scaler = StandardScaler()
train_data[['Age', 'Fare']] = scaler.fit_transform(train_data[['Age', 'Fare']])
test_data[['Age', 'Fare']] = scaler.transform(test_data[['Age', 'Fare']])
```

```python
# Convert categorical variables to numerical
label_encoder = LabelEncoder()
train_data['Sex'] = label_encoder.fit_transform(train_data['Sex'])
test_data['Sex'] = label_encoder.transform(test_data['Sex'])
train_data['Embarked'] = label_encoder.fit_transform(train_data['Embarked'])
test_data['Embarked'] = label_encoder.transform(test_data['Embarked'])
```

### 새로운 column 생성

```python
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch'] + 1
test_data['FamilySize'] = test_data['SibSp'] + test_data['Parch'] + 1
```

```python
bins = [0, 12, 18, 30, 50, 100]
labels = ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']
train_data['AgeGroup'] = pd.cut(train_data['Age'], bins=bins, labels=labels)
test_data['AgeGroup'] = pd.cut(test_data['Age'], bins=bins, labels=labels)
```

```python
bins = [-1, 7.91, 14.45, 31, 1000]
labels = ['Low', 'Medium', 'High', 'Very High']
train_data['FareGroup'] = pd.cut(train_data['Fare'], bins=bins, labels=labels)
test_data['FareGroup'] = pd.cut(test_data['Fare'], bins=bins, labels=labels)
```

### model implementation

```python
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
train_df, val_df = train_test_split(train_data, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Define the features to use for training
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

# Create the training and validation data
X_train = train_df[features]
y_train = train_df['Survived']
X_val = val_df[features]
y_val = val_df['Survived']

# Create and train the logistic regression model
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predict the survival status for the validation data
y_val_pred = lr_model.predict(X_val)

# Calculate the evaluation metrics
accuracy = accuracy_score(y_val, y_val_pred)
precision = precision_score(y_val, y_val_pred)
recall = recall_score(y_val, y_val_pred)
f1 = f1_score(y_val, y_val_pred)

# Print the evaluation metrics
print("Logistic Regression:")
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("Precision: {:.2f}%".format(precision * 100))
print("Recall: {:.2f}%".format(recall * 100))
print("F1 Score: {:.2f}%".format(f1 * 100))
```
